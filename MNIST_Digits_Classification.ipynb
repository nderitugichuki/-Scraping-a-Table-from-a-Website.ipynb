#Importing the Modules

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from sklearn.metrics import confusion_matrix

np.random.seed(0)
```

Importing the Data

```python
#loading the mnist data
from keras.datasets import mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
print(X_train.shape,y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
X_train
```

```python
plt.imshow(X_train[0], cmap='gray')
plt.show()
```

visualization

```python
#28x28 images of hand written digits 0-9
num_clases = 10
f, ax = plt.subplots(1, num_clases, figsize=(20,3))
for i in range(0, num_clases):
  sample = X_train[y_train == i][0]
  ax[i].imshow(sample, cmap='gray')
  ax[i].set_title('The number: {}'.format(i))
  ax[i].axis('off')
```

```python
for i in range(10):
  print(y_train[i])
```

```python
 #scaling the data
 X_train = X_train/255.0
 X_test = X_test/255.0
```

```python
#checking for the scaled data
plt.imshow(X_train[0],cmap='gray')
plt.show()
```

```python
#model building with keras
model=keras.models.Sequential()
model.add(keras.layers.Flatten(input_shape=[28,28]))
model.add(keras.layers.Dense(300,activation='relu'))
model.add(keras.layers.Dense(100,activation='relu'))
model.add(keras.layers.Dense(10,activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])


```

```python
#training the model
model.fit(X_train,y_train,epochs=10)
```

```python
val_loss, val_acc = model.evaluate(X_test, y_test)
print(val_loss, val_acc)
```

```python
#Get predictions
predictions=model.predict(X_test)
predicted_labels=np.argmax(predictions,axis=1)
```

```python
#Display Sample Predictions
num_samples = 10
plt.figure(figsize=(15, 5))
for i in range(num_samples):
    plt.subplot(2, 5, i+1)
    plt.imshow(X_test[i], cmap='gray')
    plt.title(f"True: {y_test[i]}\nPred: {predicted_labels[i]}")
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
#Confusion Matrix
cm = confusion_matrix(y_test, predicted_labels)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='viridis',
            xticklabels=np.arange(10),
            yticklabels=np.arange(10))
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix for MNIST Predictions')
plt.show()
```

```python
#Visualize Misclassified Examples
misclassified_indices = np.where(predicted_labels != y_test)[0]
num_misclassified = 10
plt.figure(figsize=(15, 5))
for i in range(num_misclassified):
    idx = misclassified_indices[i]
    plt.subplot(2, 5, i+1)
    plt.imshow(X_test[idx], cmap='gray')
    plt.title(f"True: {y_test[idx]}\nPred: {predicted_labels[idx]}")
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python

#Per-Class Accuracy
from sklearn.metrics import accuracy_score
per_class_accuracy = {}
for digit in range(10):
    digit_indices = np.where(y_test == digit)[0]
    digit_acc = accuracy_score(y_test[digit_indices], predicted_labels[digit_indices])
    per_class_accuracy[digit] = digit_acc
    print(f"Digit {digit}: Accuracy = {digit_acc:.4f}")
```
